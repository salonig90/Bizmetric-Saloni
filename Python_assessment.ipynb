# 14. Digitalize the book allotment process for school. Charges are mentioned here in the given table:
| STANDARD | 1st-4th | 5th-8th | 9-10th | Notebook     | 100 pages | 200 pages |
| -------- | ------- | ------- | ------ | ------------ | --------- | --------- |
| BOOKS    |         |         |        |              |           |           |
| Hindi    | 60      | 100     | 150    | square       | 40        | 70        |
| Marathi  | 60      | 100     | 150    | 4lines       | 30        | 50        |
| English  | 80      | 100     | 150    | 2lines       | 30        | 50        |
| Science  | 90      | 120     | 200    | single lines | 60        | 100       |
| Maths    | 100     | 140     | 250    | A4 notebook  | 100       | 180       |

```python

standard_books = {
    '1st-4th': {'Hindi':60, 'Marathi':60,'English':80,'Science':90,'Maths':100},
    '5th-8th': {'Hindi':100,'Marathi':100,'English':100,'Science':120,'Maths':140},
    '9th-10th':{'Hindi':150,'Marathi':150,'English':200,'Science':200,'Maths':250}
}

notebooks = {
    'square':{'100pages':40,'200pages':70},
    '4lines':{'100pages':30,'200pages':50},
    '2lines':{'100pages':30,'200pages':50},
    'singlelines':{'100pages':60,'200pages':100},
    'A4notebook':{'100pages':100,'200pages':180}
}


book_orders = {}
total = 0
while True:

    book_or_notebook = input("\nDo you want book or notebook? (type 'done' to stop): ").lower()
    if book_or_notebook == 'done':
        break

# chose which class book 

    if book_or_notebook == 'book':

        stdbook = int(input("Enter which class book (1-10): "))

        if 1 <= stdbook <= 4:
            standard = '1st-4th'
        elif 5 <= stdbook <= 8:
            standard = '5th-8th'
        elif 9 <= stdbook <= 10:
            standard = '9th-10th'
        else:
            print("Invalid class")
            continue

        print("\nAvailable books for this class:\n")

        for b in standard_books[standard]:
            print(b, ":", standard_books[standard][b])

        while True:
            book = input("\n Enter book name (type 'done' to stop): ")

            if book.lower() == 'done':
                break

            if book in standard_books[standard]:
                qty = int(input("Enter quantity: "))
                price = standard_books[standard][book]
                amount = qty * price

                book_orders[book] = [qty, price, amount]
                total += amount

                print(book, "added successfully")
            else:
                print("Book not available")

#notebook part

    elif book_or_notebook == "notebook":

        print("\nAvailable notebook types:\n")
        for n in notebooks:
            print(n)

        notebooktype = input("\nEnter notebook type: ")
        pages = input("Enter pages (100pages / 200pages): ")

        if notebooktype in notebooks and pages in notebooks[notebooktype]:
            qty = int(input("Enter quantity: "))
            price = notebooks[notebooktype][pages]
            amount = qty * price

            book_orders[notebooktype] = [qty, price, amount]
            total += amount
            
        else:
            print("Invalid notebook choice")
            continue

    else:
        print("Invalid choice")
        continue


#generate total bill
print("\n" + "-"*50)
print("|{:^50}|".format("Books BILL"))
print("-"*50)

for item, details in book_orders.items():
    print("Item:", item)
    print("Quantity:", details[0])
    print("Price:", details[1])
    print("Amount:", details[2])
    print("-"*50)

print("TOTAL AMOUNT =", total)

```

```python
# # 35. List2 = [2,70,'work', para, 2.5, [1,2,3], (1,2), {1,2}, {1:'a', 2:'b'}, 3,10,302.5]
# # Multiply each and every element by 2 and display the answer

List2 = [2 , 70 , 'work' , 'para' , 2.5, [1,2,3] , (1,2), {1,2} , {1:'a', 2:'b'}, 3, 10 , 302.5 ]
l1= [] 
for i in List2:
    if type(i) in [str,int,float,list,tuple]:
        l1.append(i*2)
    elif type(i) == set:
        # l1.append({x*2 for x in i})
        new_set = set()
        for x in i:
            new_set.add(x*2)
        l1.append(new_set)
    elif type(i) == dict:
        new_dict = {}
        for key in i:
            new_dict[key] = i[key] * 2
        l1.append(new_dict)
print(l1)
```
