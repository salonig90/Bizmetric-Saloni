**List Comprehension**

```python
# 1.Write a list comprehension to generate squares of numbers from 1 to 10.
lc_square = [x*x for x in range(1,11)]
print(lc_square)
```

```python
#2. Create a list of even numbers between 1 and 50 using list comprehension.
lc_even_odd = [x for x in range (1,51) if x%2==0 ]
print(lc_even_odd)
```

```python
# 3. Convert all strings in a list to uppercase using list comprehension.
li = ['Hello','all']
string_upper = [x.upper()  for x in li]
print(string_upper)
```

```python
# 4. Given a list of integers, create a new list that contains only the positive numbers.
li= [-2,2,5,8,-9]
newlist_lc = [x for x in li if x>0]
print(newlist_lc)
```

```python
# 5. Create a list of tuples (num, num^2) for numbers 1 to 5.

tuple_lc = [ (x,x**2) for x in range(1,6)]
print(tuple_lc)
```

```python
# 6. Extract all vowels from a given string using list comprehension.

st = 'hello'
lc_vowel = [ x for x in st if x in ('aeiou')]
print(lc_vowel)
```

```python
# 7. Flatten a 2D list using list comprehension.
nums = [[1, 2], [3, 4]]

flat = [n for sub in nums for n in sub]
print(flat)
```

```python
# 8. Replace all negative numbers in a list with 0 using list comprehension.
li = [ -1,2,-3,-7,9]
replace_lc = [0 if n < 0 else n for n in li]
print(replace_lc)
```

```python
# 9. Given a list of words, create a list of lengths of each word. 

li = ['Hello','all']
len_lc = [ len(sub) for sub in li]
print(len_lc)
```

```python
# 10. Filter out words that start with the letter 'A' or 'a'. 
li = ['apple', 'lemon', 'app']
lc_a = [x for x in li if x.lower().startswith('a')]  #here x.lower(0 is convert every letter in small so A and a are treated the same)
print(lc_a)
```

```python
# 11. From a list of numbers, generate a list of “even” or “odd” strings using list comprehension.
# (Like → [“even”, “odd”, “odd”, “even”…])
li = [2,3,4,5,6,7,8]
stri_lc = ['even' if x%2 == 0 else 'odd' for x in li ]
print(stri_lc)
```

```python
# 12. Create a list of numbers divisible by both 3 and 5 in range 1–100.

li = [ x for x in range(1,101) if x%3 ==0 and x%5 == 0]
print(li)
```

```python
# 13. Write a nested list comprehension to generate a multiplication table  for 1–5.
table_lc = [[i*j for j in range(1,6)] for i in range (1,6)]
print(table_lc) 

a = 1
b = 'hello'
a 
```

```python
# 14. Convert a dictionary’s keys into a list using list comprehension.
dic = {'a' : 1 , 'b' : 2 } 
newli = [ key for key in dic ] 
print(newli)
```

```python
# 15. Extract numeric digits from a string using list comprehension.
st = 'sldk324f5'
newli = [digit for digit in st if digit.isdigit()]
print(newli)
```

```python
# 16. Use list comprehension to remove all spaces from a string.
st = 'hello all myself sallu'
result = ''.join([ch for ch in st if ch != ' ']) 
result
```

```python
# 17. Create a list of characters that appear more than once in a string.

st = 'aabbcc'
result = [ s for s in set(st) if st.count(s) > 1 ]
result
```

```python
# 18. From a list of sentences, generate a list of all words (split using list comprehension).
li = [ 'hello world how is the universe' ]
res = [x for sub in li for x in sub.split() ] 
res
```

```python
# 19. Create a list of unique elements from a list using list comprehension + condition.
li = [ 2, 4 , 3, 2, 4 ]
res = [x for x in set(li)]
res
```

```python
# 20. Generate all pairs (x, y) where x is from list A and y is from list B 
# (cartesian product).
A = [1,2,3]
B = [4,5,6]
lc_res = [(x,y) for x in A for y in B ]
lc_res
```
