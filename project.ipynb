```python
class DatabaseConnection:
    """Handles database connection."""

    def createconnection(self):
        """Create SQL Server connection."""
        try:
            import pyodbc
            import sys
            self.conn1 = pyodbc.connect(
                r'Driver=ODBC Driver 18 for SQL Server;'
                r'Server=DESKTOP-CGL04K2\SQLEXPRESS;'
                r'Database=Bill;'
                r'Trusted_Connection=yes;'
                r'Encrypt=no;'
            )
            print("Database Connected Successfully")

        except:
            print(sys.exec_info())


# MENU & ORDER CLASS 

class MenuOrder(DatabaseConnection):
    """Handles menu display and order input."""

    def DisplayMenuTakeOrder(self):
        """Fetch menu and store orders."""
        self.cursor = self.conn1.cursor()       # obj to run sql queries
        self.cursor.execute('SELECT * FROM MenuItem')    #fetching orders 
        self.records = self.cursor.fetchall()   #storing all rows into records 

        print("\n---- MENU ----")
        for row in self.records:
            print(row)

        self.store_item = {}

        while True:
            
            item = input("Enter the item: ")
            if item == "done":
                break

            for row in self.records:
                menu_name = row[1]
                price = row[2]

                if item == menu_name:
                    qty = int(input("Enter quantity: "))
                    total = qty * price

                    self.store_item[menu_name] = {
                        "qty": qty,
                        "price": price,
                        "total": total
                    }

        print("\nOrdered Items:", self.store_item)

# ORDER DATABASE CLASS 

class OrderDB(MenuOrder):
    """Stores order details into database."""

    def OrderDisplay(self):
        """Insert orders into Orders table."""
        try:
            for item_name, details in self.store_item.items():

                quantity = details["qty"]
                price = details["price"]
                total = details["total"]

                self.cursor.execute(f"""
                    INSERT INTO Orders (item, quantity, price, total_price, creationdate)
                    VALUES ('{item_name}', {quantity}, {price}, {total}, GETDATE())
                """)

            self.conn1.commit()
            print("Order inserted successfully!")

        except Exception as e:
            print("Error:", e)


#BILLING CLASS

class Billing(OrderDB):
    """Generates and saves bill."""

    def GenerateBill(self):
        """Display bill on console."""

        from datetime import datetime
        now = datetime.now()
        dt_string = now.strftime("%d-%m-%Y %H:%M:%S")

        print('-'*90)
        print("|{:^90}|".format('RadhaKrishn Hotel - BILL'))
        print("|{:^90}|".format(f'Date & Time: {dt_string}'))
        print('-'*90)
        print('|{:<10} {:<25} {:<15} {:<15} {:>20}|'.format("Sr","Menu","Quantity","Price","Amount"))
        print('-'*90)

        sr_no = 1
        self.total_cost = 0

        for item_name, details in self.store_item.items():
            qty = details['qty']
            price = details['price']
            amount = details['total']

            self.total_cost += amount

            print('|{:<10} {:<25} {:<15} {:<15} {:>20}|'.format(sr_no, item_name, qty, price, amount))
            sr_no += 1

        print('-'*90)
        print('|{:<65}{:>25}|'.format("Total Bill", self.total_cost))
        print('-'*90)


    def PrintBillFile(self):
        """Save bill into text file."""

        from datetime import datetime
        now = datetime.now()
        dt_string = now.strftime("%d-%m-%Y %H:%M:%S")

        file = open("Bill.txt", "w")

        file.write('-'*90 + "\n")
        file.write("|{:^88}|\n".format('Welcome to RadhaKrishn Hotel'))
        file.write("|{:^88}|\n".format(f'Date & Time: {dt_string}'))
        file.write('-'*90 + "\n")
        file.write('|{:<20} {:<20} {:<20} {:>25}|\n'.format("Sr","Menu","quantity","price"))
        file.write('-'*90 + "\n")

        sr_no = 1
        total_cost = 0

        for item_name, details in self.store_item.items():
            qunt = details["qty"]
            total_per_item = details["total"]

            total_cost += total_per_item

            row = '|{:<20} {:<20} {:<20} {:>25}|'.format(sr_no, item_name, qunt, total_per_item)
            file.write(row + "\n")
            sr_no += 1

        file.write('-'*90 + "\n")
        total_line = '|{:<60}{:>28}|'.format("Total", total_cost)
        file.write(total_line + "\n")
        file.write('-'*90 + "\n")

        file.close()

        print("\nBill saved successfully in 'Bill.txt'")
```

```python
class DatabaseConnection:
    """Handles database connection."""

    def createconnection(self):
        try:
            import pyodbc
            import sys

            self.conn1 = pyodbc.connect(
                r'Driver=ODBC Driver 18 for SQL Server;'
                r'Server=DESKTOP-CGL04K2\SQLEXPRESS;'
                r'Database=CollegeDB;'   # changed here
                r'Trusted_Connection=yes;'
                r'Encrypt=no;'
            )
            print("Database Connected Successfully")

        except:
            print(sys.exec_info())

class CollegeInput(DatabaseConnection):
    
    def __init__(self):
        self.subjects = ['HR', 'Finance', 'Marketing', 'DS']
        self.l1_fees = 200000
        self.analytics_fee = self.l1_fees + (self.l1_fees * 0.10)
        self.hostel_fee = 200000
        self.food_per_month = 2000
        self.transport_charge = 13000

    def take_input(self):

        self.subject = input("Enter your subject: ")
        self.analytics_opt = input("Want analytics? (Y/N): ").upper()
        self.hostel_opt = input("Want hostel? (Y/N): ").upper()
        self.food_months = int(input("Enter food months: "))
        self.transportation = input("semester/annual: ").lower()

    def calculate_fees(self):

        # course fee
        if self.subject != 'DS' and self.analytics_opt == 'Y':
            self.course_fee = self.analytics_fee
        else:
            self.course_fee = self.l1_fees

        # hostel
        if self.hostel_opt == 'Y':
            self.hostel_cost = self.hostel_fee
        else:
            self.hostel_cost = 0

        # food
        self.food_cost = self.food_months * self.food_per_month

        # transport
        if self.transportation == 'semester':
            self.transport_cost = self.transport_charge * 2
        else:
            self.transport_cost = self.transport_charge

        # total
        self.total_cost = self.course_fee + self.hostel_cost + self.food_cost + self.transport_cost

        print("Total Fee:", self.total_cost) 

class FeeDB(CollegeInput):

    def store_fee(self):
        try:
            self.cursor = self.conn1.cursor()

            self.cursor.execute(f"""
                INSERT INTO CollegeFees 
                (subject, analytics_opt, hostel_opt, food_months, transportation,
                 course_fee, hostel_fee, food_fee, transport_fee, total_fee)
                VALUES 
                ('{self.subject}', '{self.analytics_opt}', '{self.hostel_opt}',
                 {self.food_months}, '{self.transportation}',
                 {self.course_fee}, {self.hostel_cost}, {self.food_cost},
                 {self.transport_cost}, {self.total_cost})
            """)

            self.conn1.commit()
            print("Fee details inserted successfully!")

        except Exception as e:
            print("Error:", e)

class FeeBilling(FeeDB):
    
    def generate_bill(self):
        
        from datetime import datetime
        now = datetime.now()
        dt_string = now.strftime("%d-%m-%Y %H:%M:%S")

        print("-"*82)
        print("|{:^80}|".format("College Fee Receipt"))
        print("|{:^80}|".format(f"Date & Time: {dt_string}"))
        print("-"*82)

        print("| {:<80}|".format(f"Subject           : {self.subject}"))
        print("| {:<80}|".format(f"Analytics Opted   : {self.analytics_opt}"))
        print("| {:<80}|".format(f"Hostel Opted      : {self.hostel_opt}"))
        print("| {:<80}|".format(f"Food Months       : {self.food_months}"))
        print("| {:<80}|".format(f"Transportation    : {self.transportation}"))

        print("-"*82)

        print("| {:<80}|".format(f"Course Fee        : {self.course_fee}"))
        print("| {:<80}|".format(f"Hostel Fee        : {self.hostel_cost}"))
        print("| {:<80}|".format(f"Food Fee          : {self.food_cost}"))
        print("| {:<80}|".format(f"Transport Fee     : {self.transport_cost}"))

        print("-"*82)
        print("| {:<80}|".format(f"TOTAL FEE         : {self.total_cost}"))
        print("-"*82)

```
