```python
#5.
#student validation 
import re
import os

FILE_NAME = "studentrecords1.txt"
records = {}

# VALIDATION FUNCTIONS  

def ValidName(name):
    if not name:
        raise ValueError("Name cannot be empty")
    if not name.isalpha():
        raise ValueError("Name must contain alphabets only")
    return name


def ValidMobileNo(mobile):
    if not mobile.isdigit() or len(mobile) != 10:
        raise ValueError("Mobile must contain exactly 10 digits")
    return mobile


def ValidatePan(pan):
    if not re.match(r"[A-Z]{5}[0-9]{4}[A-Z]$", pan):
        raise ValueError("Invalid PAN number format")
    return pan


def ValidateMarks(marks10, marks12, graduation_marks):
    if not (0 <= marks10 <= 100 and
            0 <= marks12 <= 100 and
            0 <= graduation_marks <= 100):
        raise ValueError("Marks must be between 0 and 100")
    return marks10, marks12, graduation_marks


# CHECK DUPLICATE PAN 

def pan_exists(pan):

    # Check in current session
    for record in records.values():
        if record["pan"] == pan:
            return True

    # Check in file
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, "r") as file:
            for line in file:
                if pan in line:
                    return True

    return False


#GET NEXT ID 

def get_next_id():
    if not os.path.exists(FILE_NAME):
        return 1

    with open(FILE_NAME, "r") as file:
        lines = file.readlines()
        return len(lines) + 1


# SAVE RECORD TO FILE 

def save_to_file(emp_id, data):
    with open(FILE_NAME, "a") as file:
        file.write(
            f"Emp id: {emp_id}, "
            f"Name: {data['name']}, "
            f"Mobile: {data['mobile']}, "
            f"PAN: {data['pan']}, "
            f"10th: {data['10th']}, "
            f"12th: {data['12th']}, "
            f"Graduation: {data['graduation']}\n"
        )


#MAIN FUNCTION 

def create_user():

    while True:
        try:
            name = ValidName(input("Enter name: "))
            mobile = ValidMobileNo(input("Enter mobile number: "))
            pan = ValidatePan(input("Enter PAN: "))

            if pan_exists(pan):
                print("PAN already exists. Record not added.")
                continue

            marks10 = float(input("Enter 10th marks: "))
            marks12 = float(input("Enter 12th marks: "))
            graduation = float(input("Enter Graduation marks: "))

            ValidateMarks(marks10, marks12, graduation)

            emp_id = get_next_id()

            user_dict = {
                "name": name,
                "mobile": mobile,
                "pan": pan,
                "10th": marks10,
                "12th": marks12,
                "graduation": graduation
            }

            save_to_file(emp_id, user_dict)

            print("\nRecord Saved Successfully")
            print("Generated Employee ID:", emp_id)

        except Exception as e:
            print("Error:", e)

        choice = input("\nDo you want to enter another record? (Y/N): ").strip().upper()
        if choice != "Y":
            print("Exiting program...")
            break


create_user()
```

```python
class signin():
    def __init__(self):
        self.username = input("Enter username:")
        self.password = input("Enter password:")
    
    def showsignin(self):
        if self.username in registered_user:
            if registered_user[self.username] == self.password:
                print(f"Successfully signed in by {self.username}")
            else:
                print("Incorrect password")
        else:
            print("user not registered")
    

signinobj = signin()
signinobj.showsignin()
```

```python
class login(signin):
    def __init__(self):
        super().__init__()   # calls parent class __init__ to take username & password

    def create_account(self):
        if self.username in registered_user:
            print("Username already exists")
        else:
            registered_user[self.username] = self.password
            print("Account created successfully")
            print("Updated users:", registered_user)

loginobj = login()
loginobj.create_account()
```
